```python
import pandas as pd
import matplotlib.pyplot as plt

# pour lire le fichier
df = pd.read_excel("data-sell4all.xlsx")
df.head()
df.info()
# bach nhydo espace
df.columns = df.columns.str.strip()

# convertiw age et depenses
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  
df['Dépenses des clients'] = pd.to_numeric(df['Dépenses des clients'], errors='coerce')  

df['Age'] = df['Age'].fillna(0).astype(int)

# pour calculer  la médiane et la moyenne 
median_age = df['Age'].median()
mean_age = df['Age'].mean()
median_depense = df['Dépenses des clients'].median()
mean_depense = df['Dépenses des clients'].mean()
# nombre des lignes et colummes
nbr_lignes = df.shape[0]
nbr_colone = df.shape[1]
# les types des types_donnes
types_donnes = df.dtypes

print("les nombres des lignes est " , nbr_lignes)
print("les nombres des colones est " , nbr_colone)
print("les types des donnes est : " , types_donnes)
print("Mediane de la colonne age est : " , median_age)
print("Moyenne de la colonne age est : " , mean_age)
print("Mediane de la colonne Dépenses des clients est : " , median_depense)
print("Moyenne de la colonne Dépenses des clients est : " , mean_depense)

# une visualisation des données du graphique
plt.figure(figsize=(12, 6))
df.groupby('Pays')['Dépenses des clients'].sum().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Dépenses des clients par pays')
plt.xlabel('Pays')
plt.ylabel('Dépenses des clients')
plt.xticks(rotation=60)

#pour supprimer les doublons
df = df[df['Dépenses des clients'] >= 10]
df = df.drop_duplicates()

colummes_selected = ["Pays" , "Age" , "Genre" , "Dépenses des clients"]
df_nettoyer = df[colummes_selected]
df_nettoyer.to_csv("nouveau_fichier.csv")

print("les donnes nettoyées ont ete écrites dans le fichier nouveau_fichier.csv")
# pour afficher le graphe
# plt.show()

```
